immigration data analysis

```python
df1 = pd.read_excel('8Canada.xlsx', sheet_name=1,skiprows=20, skipfooter=2)
df1.head(2)
```

- rename columns
- drop columns 
- add columns

rename

```python
cols_to_rename = {
    'OdName':'Country',
    'AreaName':'Continent',
    'RegName':'Region',
    'DevName':'Status'
    }
df1=df1.rename(columns=cols_to_rename)
df1.head(2)
```

drop columns

```python
# check no of nunique values in data
print("Type unique:", df1.Type.nunique())
print("Coverage Unique value:",df1.Coverage.nunique())

# count the unique value in a column
print("Coverage Unique values count\n",df1.Coverage.value_counts())
```

```python
cols_to_drop=["AREA","REG","DEV","Type","Coverage"]
df1=df1.drop(columns=cols_to_drop)
df1.head(2)
```

set index= not for every dataframe you use

```python
df1= df1.set_index('Country')
df1.head()
```

make al column to lower case- optional step

```python
df1.columns=[str(name).lower() for name in df1.columns.tolist()]
df1.head()
```

```python
years = list(map(str, range(1980,2014)))
print(years)
```

adding a column(total)

```python
df1[years].sum(axis=1)   # axis 1 for row wise sum and axis 0 fro columnwise sum
```

```python
df1['total']=df1[years].sum(axis=1)
df1.head()
```

```python
df1.isnull().sum().plot(figsize=(10,1),kind='bar')
```

visualization

total immigration visualization

```python
df1.total.plot(rot=90)
```

```python
df1.total.plot(rot=90,
               figsize=(25,5),
               xticks=range(0,df1.shape[0],2))
```

```python
df1.total.plot(rot=90,
               figsize=(25,5),
               xticks=range(0,df1.shape[0],2),
               title='Total Immigration')
```

```python
# top 25 country with highest immigration

df1.sort_values('total',ascending=False)
```

```python
df1.sort_values('total',ascending=False).head(25).plot(kind='bar',
                                                       y='total', figsize=(15,5))
```

```python
df1.sort_values('total',ascending=False).head(25).plot(kind='bar',
                                                       y='total', figsize=(15,5),color='red',
                                                       alpha=.5,title='Top 25 countries with highest immigration',
                                                       xlabel='Country',ylabel='Immigration')
```
