```python
from dputils.scrape import Tag, Scraper
```

```python
url="https://timesofindia.indiatimes.com/news"
```

```python
timeofindian=Scraper(url)
```

```python
target = Tag('ul', cls ="HytnJ")
items = Tag('li')
title = Tag("p", cls="CRKrj")
desc = Tag("p", cls ="W4Hjm")
```

```python
import pandas as pd
import numpy as np
```

```python
result=timeofindian.get_repeating_page_data(target,items,heading=title, desc=desc)
```

```python
df=pd.DataFrame(result)
df
```
